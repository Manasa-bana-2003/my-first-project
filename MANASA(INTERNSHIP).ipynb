```python
Bedrooms=[];Bathrooms=[]
Location=[];Price=[]
Size=[];Status=[]
Facing=[];Type=[]
property_types=['Apartment','Builder Floor','Villa','Residential Plot','Independent House','Independent Floor','Studio Apartment']
for i in range(1,54):
    url=f'https://www.makaan.com/hyderabad/hyderabad-south-property-10238?property_types=apartment,builder-floor,villa,residential-plot,independent-house,studio-apartment&page={i}'
    print(f'Scrapping Info from Page-->{i}')
    clear_output(wait=True)
    time.sleep(0.01)#waiting time to move from one place to other
    d=requests.get(url)
    soup=BeautifulSoup(d.content,'html')

    a=soup.find_all('a',attrs={'class':'typelink'})
    for i in a:
        i=i.span.text #you need only number which is present in start
        i=i.replace('Residential Plot','0').replace(' ','')
        Bedrooms.append(i)
    b=soup.find_all('span',attrs={ 'itemprop':"addressLocality"})
    for i in b:
        Location.append(i.text)
    c=soup.find_all('div',attrs={'data-type':"price-link"})
    for i in c:
        i=i.text.strip()
        if "Cr" in i:
            i=i.replace(" Cr",'')#Replacing with empty string
            i=float(i)*100 #converting intlo lakhs
        else:
            i=i.replace(" L","")
            i=float(i)
        Price.append(i)
    d=soup.find_all('td',attrs={'class':"size"})
    for i in d:
        i=i.text.strip()
        Size.append(i)
    e=soup.find_all('td',attrs={'class':"val"})
    for i in e:
        i=i.text
        Status.append(i)
    f=soup.find_all('div',attrs={'class':'title-line'})
    for i in f:
        i=i.text
        bathroom_count=re.findall(r'(\d) Bathrooms',i)
        if bathroom_count:
            Bathrooms.append(int(bathroom_count[0]))
        else:
            Bathrooms.append(0)
        facing_directions=re.findall(r',(North|South|East|West|NorthEast|NorthWest|SouthEast|SouthWest)facing',i)
        if facing_directions:
            Facing.append(facing_directions[0])
        else:
            Facing.append('None')
    g=soup.find_all('div',attrs={'class':'title-line'})
    for i in g:
        i=i.text
        for p_type in property_types:
            if p_type in i:
                Type.append(p_type)
        
```

```python
#Removing outliers from the dataframe -->we will create a final dataframe
def remove_outliers(df,threshold=3):
    df_cleaned=df.copy()#copy of dataframe
    
    #The threshold value determines how for the first and third quadrant
    for col in df_cleaned.select_dtypes(include=['int64','float64']).columns:
        Q1=df_cleaned[col].quantile(0.25) #25th percentile
        Q3=df_cleaned[col].quantile(0.75) #75th percentile
        IQR=Q3-Q1 #InterQuantile Range
        lower_bound=Q1-threshold*IQR
        upper_bound=Q3+threshold*IQR
        
        df_cleaned=df_cleaned[(df_cleaned[col] >=lower_bound) &
                              (df_cleaned[col]<= upper_bound)
                             ]
    return df_cleaned
```
